# Grouping and Summarizing Dataframes

```python
#Group the data 'df' by 'month' and 'day' and find the mean value for column 'rain' and 'wind'.
df = pd.read_csv('https://query.data.world/s/vBDCsoHCytUSLKkLvq851k2b8JOCkF')
#df.head(20)
df_by_month_day = df.groupby(['month', 'day'])
df_1 = df_by_month_day['rain'].sum().sort_values(ascending = False)
#df_1 = df_by_month_day.rain.sum().sort_values(ascending = False)
# df_1 = df_by_month_day['rain'].mean()
print(df_1.head(20))
```

```python
#calculate rain across each month as percentage
(df.groupby('month').rain.sum() / sum(df['rain'])) * 100
```

```python
#calculate wind across each month as percentage
(df.groupby('month').wind.sum() / sum(df['wind'])) * 100
```

```python
df_by_month_day.describe()
```

```python
#Dataframe grouping
#Group the data 'df' by 'month' and 'day' and find the mean value for column 'rain' and 'wind'.
df_by_month_day = df.groupby(['month', 'day'])
df_1 = df_by_month_day['rain', 'wind'].mean()
print(df_1.head(20))
```

# Lambda function & Pivot tables

```python
df = pd.read_csv('https://query.data.world/s/vBDCsoHCytUSLKkLvq851k2b8JOCkF')
df.head()
```

```python
#create a function to be applied
def is_happening(x):
    return x > 0
#create new columns
df['is_raining'] = df['rain'].apply(is_happening)
df['is_wind'] = df['wind'].apply(is_happening)
df.head()
```

```python
#lambda way of the above
df['is_raining'] = df['rain'].apply(lambda x: x > 0)
df['is_wind'] = df['wind'].apply(lambda x: x > 0)
df.head()
```

```python
by_month = df.groupby('month')
by_month.is_raining.mean()
```

```python
#Create a new column 'XY' which consist of values obtained from multiplying column 'X' and column 'Y'.
df = pd.read_csv('https://query.data.world/s/vBDCsoHCytUSLKkLvq851k2b8JOCkF')
df['XY'] = df['X'] * df['Y']
print(df.head(20))
```

```python
#Group the data 'df' by 'month' and 'day' and find the mean value for column 'rain' and 'wind' using the pivot table command.
df = pd.read_csv('https://query.data.world/s/vBDCsoHCytUSLKkLvq851k2b8JOCkF')
df_1 = df.pivot_table(values=['rain', 'wind'], index=['month','day'], aggfunc = 'mean')
print(df_1.head(20))
```
